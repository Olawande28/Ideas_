```python
def first_diff(s1, s2):
  for i in range(min(len(s1), len(s2))):
    if s1[i] != s2[i]:
      return i
  return -1
print(first_diff('abcdef', 'abcde'))
print(first_diff('abc', 'aBc'))
```

```python
import math
def binom(n, k):
  if k > n or k < 0:
    return 0
  return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
print(binom(5, 2))
```

```python
import random
def random_n(n):
  if n < 0:
    raise ValueError('n must be non-negative')
  start = 10 ** (n - 1)
  end = 10 ** n - 1
  return random.randint(start, end)
print(random_n(3))
```

```python
def number_of_factors(n):
  count = 0
  for i in range(1, n + 1):
    if n % i == 0:
      count += 1
  return count
print(number_of_factors(12))
```

```python
def factors(n):
  return [i for i in range(1, n + 1) if n % i == 0]
print(factors(12))
```

```python
def closest(L, n):
  filtered = [x for x in L if x <= n]
  return max(filtered) if filtered else None
print(closest([1, 2, 3, 4, 5], 3))
```

```python
def matches(s1, s2):
  min_length = min(len(s1), len(s2))
  match_count = 0
  for i in range(min_length):
    if s1[i] == s2[i]:
      match_count += 1
  return match_count
result = matches('abcde', 'abcd')
print(result)
```

```python
def findall(s, char):
  if len(char) != 1:
    raise ValueError
  positions = []
  for index, current_char in enumerate(s):
    if current_char == char:
      positions.append(index)
  return positions
result = findall('abcabcabc', 'c')
print(result)
#
```

```python
def one_away(str1, str2):
    if len(str1) != len(str2):
        return False
    diff_count = 0
    for char1, char2 in zip(str1, str2):
        if char1 != char2:
            diff_count += 1
            if diff_count > 1:
                return False
    return diff_count == 1  # Return True if exactly one difference

# Example usage
print(one_away("bike", "hike"))
print(one_away("hello", "hallo"))
print(one_away("test", "tent"))
print(one_away("abcd", "abcde"))
print(one_away("abcd", "abcd"))
print(one_away("abc", "xyz"))
```

```python
def is_prime(num):
  if num < 2:
    return False
  for i in range(2, int(num ** 0.5) + 1):
    if num % i == 0:
      return False
  return True
def primes(n = 100, start = 2):
  prime_list = []
  current_number = start
  while len(prime_list) < n:
    if is_prime(current_number):
      prime_list.append(current_number)
    current_number += 1
  return prime_list
print(primes())
print(primes(10, 20))
#
```

```python
def base_20(num):
  if num == 0:
    return 'A'
  digits = []
  letters = 'ABCDEFGHIJKLMNOPQRST'
  while num > 0:
    remainder = num % 20
    digits.append(letters[remainder])
    num //= 20
  return ''.join(reversed(digits))
print(base_20(123456789))
#
```

```python
def merge_with_sort(l1, l2):
  l3 = l1 + l2
  l3.sort()
  return l3
print(merge_with_sort([1, 2, 3, 7, 9], [3, 4, 5, 6, 8]))
#
```

```python
def merge_without_sort(l1, l2):
  merged_list = []
  x,y = 0,0
  while x < len(l1) and y < len(l2):
    if l1[x] < l2[y]:
      merged_list.append(l1[x])
      x += 1
    else:
      merged_list.append(l2[y])
      y += 1
  while x < len(l1):
    merged_list.append(l1[x])
    x += 1
  while y < len(l2):
    merged_list.append(l2[y])
    y += 1
  return merged_list
l1 = [1,3,5,7]
l2 = [2,4,6,8]
print(merge_without_sort(l1, l2))
```

```python
def binary_search(words, w):
  left, right = 0, len(words) - 1
  while left <= right:
    mid = (left + right) // 2
    mid_word = words[mid]
    if mid_word == w:
      return True
    elif mid_word < w:
      left = mid + 1
    else:
      right = mid - 1
  return False
words = ['apple', 'banana', 'cherry', 'date', 'elderberry']
print(binary_search(words, 'cherry'))
```

```python
def check_winner(board):
  for i in range(3):
    if board[i][0] == board[i][1] == board[i][2] != ' ':
      return True
    if board[0][i] == board[1][i] == board[2][i] != ' ':
      return True
  if board[0][0] == board[1][1] == board[2][2] != ' ':
    return True
  if board[0][2] == board[1][1] == board[2][0] != ' ':
    return True
  return False
board = [['X', 'O', 'X'],
           ['O', 'X', 'O'],
           ['X', 'O', 'X']]
print(check_winner(board))
```

```python
def is_valid_sudoku(board):
  for row in board:
    if not is_valid_unit(row):
      return False

  for col in range(9):
    column = [board[row][col] for row in range(9)]
    if not is_valid_unit(column):
      return False

  for block_row in range(3):
    for block_col in range(3):
      block = []
      for row in range(3):
        for col in range(3):
          block.append(board[block_row * 3 + row][block_col * 3 + col])
      if not is_valid_unit(block):
        return False

  return True
def is_valid_unit(unit):
  unit = [x for x in unit if x != ' ']
  return len(unit) == len(set(unit)) and all(1 <= int(x) <= 9 for x in unit)
sudoku_board = [
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
]

print(is_valid_sudoku(sudoku_board))  # Should return True for a valid Sudoku

invalid_sudoku_board = [
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]  # Invalid row (repeated 9)
]

print(is_valid_sudoku(invalid_sudoku_board))
```
