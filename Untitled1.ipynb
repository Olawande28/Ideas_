```python
class Example:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def add(self):
    return self.a + self.b
  def sub(self):
    return self.a - self.b

e = Example(10,20)
print(e.add())
print(e.sub())
```

```python
from string import punctuation
class Analyzer:
  def __init__(self, s):
    for c in punctuation:
      s = s.replace(c, '')
      s = s.lower()
      self.words = s.split()
  def count_words(self):
    return len(self.words)
  def count_unique_words(self):
    return len(set(self.words))
  def starts_with(self, s):
    return len([w for w in self.words if w[:len(s)]==s])
  def number_with_length(self, n):
    return len([w for w in self.words if len(w)==n])

s = 'This is a class test'
analyzer = Analyzer(s)
print(analyzer.words)
print('Number of words: ', analyzer.count_words())
print('Number of unique words: ', analyzer.count_unique_words())
print('Number of words starting with "t": ', analyzer.starts_with('t'))
print('Number of words with length 3: ', analyzer.number_with_length(5))

```

```python
class Parent:
  def __init__(self, a):
    self.a = a
  def method1(self):
    return self.a*2
  def method2(self):
    return self.a+'!!!'
class Child(Parent):
  def __init__(self, a, b):
    self.a = a
    self.b = b
  def method1(self):
    return self.a*7
  def method3(self):
    return self.a + self.b
p = Parent('hi')
c = Child('hi', 'bye')
print('Parent method 1: ', p.method1())
print('Parent method 2: ', p.method2())
print()
print('Child method 1: ', c.method1())
print('Child method 2: ', c.method2())
print('Child method 3: ', c.method3())
```
